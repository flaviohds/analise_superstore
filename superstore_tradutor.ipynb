```python
import pandas as pd

# Carregar o conjunto de dados do GitHub
df = pd.read_csv('https://raw.githubusercontent.com/flaviohds/analise_superstore/refs/heads/main/superstore_sales.csv')

# Lista de colunas para verificar valores únicos
columns_to_check = ["Category", "Sub-Category", "Region", "Segment", "Ship Mode"]

for col in columns_to_check:
    print(f"\nObservações únicas em '{col}':")
    display(df[col].unique())
```

```python
# Dicionário de tradução para a coluna 'Category'
category_translation = {
    'Furniture': 'Móveis',
    'Office Supplies': 'Escritório',
    'Technology': 'Tecnologia'
}

# Dicionário de tradução para a coluna 'Sub-Category'
subcategory_translation = {
    'Bookcases': 'Estantes',
    'Chairs': 'Cadeiras',
    'Labels': 'Etiquetas',
    'Tables': 'Mesas',
    'Storage': 'Armazenamento',
    'Furnishings': 'Mobiliário',
    'Art': 'Arte',
    'Phones': 'Telefones',
    'Binders': 'Fichários',
    'Appliances': 'Aparelhos',
    'Paper': 'Papel',
    'Accessories': 'Acessórios',
    'Envelopes': 'Envelopes',
    'Fasteners': 'Fixadores',
    'Supplies': 'Utensílios',
    'Machines': 'Máquinas',
    'Copiers': 'Fotocopiadoras'
}

# Dicionário de tradução para a coluna 'Region'
region_translation = {
    'South': 'Sul',
    'West': 'Oeste',
    'Central': 'Central',
    'East': 'Leste'
}

# Dicionário de tradução para a coluna 'Segment'
segment_translation = {
    'Consumer': 'Consumidor',
    'Corporate': 'Corporativo',
    'Home Office': 'Home Office'
}

# Dicionário de tradução para a coluna 'Ship Mode'
ship_mode_translation = {
    'Second Class': 'Segunda Classe',
    'Standard Class': 'Classe Padrão',
    'First Class': 'Primeira Classe',
    'Same Day': 'Mesmo Dia'
}

# Aplicar as traduções ao DataFrame
df['Category'] = df['Category'].replace(category_translation)
df['Sub-Category'] = df['Sub-Category'].replace(subcategory_translation)
df['Region'] = df['Region'].replace(region_translation)
df['Segment'] = df['Segment'].replace(segment_translation)
df['Ship Mode'] = df['Ship Mode'].replace(ship_mode_translation)

print("Colunas traduzidas com sucesso!")
```

```python
# Exibir as primeiras 5 linhas do DataFrame com as traduções aplicadas
display(df.head())
```

```python
# Salvar o DataFrame traduzido para um novo arquivo CSV
df.to_csv('superstore_sales_traduzido.csv', index=False)

print("DataFrame salvo com sucesso como 'superstore_sales_traduzido.csv'!")
```
